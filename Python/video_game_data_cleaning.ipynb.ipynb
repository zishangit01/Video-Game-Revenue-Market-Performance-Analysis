```python
import pandas as pd

df = pd.read_csv(
    "video_game_final_dataset.csv",
    encoding="utf-8-sig"   # ðŸ”¥ BOM issue yahin fix ho jaata hai
)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
def convert_k(val):
    if pd.isna(val):
        return 0
    val = str(val)
    if val.endswith('K'):
        return float(val[:-1]) * 1000
    return float(val)
```

```python
df['Times Listed'] = df['Times Listed'].apply(convert_k).astype(int)
df['Number of Reviews'] = df['Number of Reviews'].apply(convert_k).astype(int)
```

```python
df['Release Date'] = pd.to_datetime(
    df['Release Date'],
    dayfirst=True,
    errors='coerce'
).dt.strftime('%Y-%m-%d')
```

```python
numeric_cols = [
    'Rating', 'Plays', 'Playing',
    'Backlogs', 'Wishlist',
    'NA_Sales', 'EU_Sales',
    'JP_Sales', 'Other_Sales',
    'Global_Sales'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
```
